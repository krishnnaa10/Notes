# SEABORN

**Seaborn** is a Python library for data visualization built on top of Matplotlib. It makes creating beautiful and informative graphs easy, especially for working with datasets.

### Graphs in Seaborn

+ Scatter Plot  
+ Line Plot  
+ Bar Plot  
+ Count Plot  
+ Box Plot  
+ Violin Plot  
+ Histogram  
+ Heatmap  
+ Pair Plot  
+ Joint Plot  


### dependency of seaborn

Seaborn depends on several Python libraries to function properly like NumPy, Pandas, Matplotlib, Scipy. Make sure these libraries are installed when using Seaborn. 

### Difference between Matplotlib and Seaborn

**Matplotlib** focuses on creating basic visualizations with full customization but requires more effort for handling complex data ranges, while **Seaborn** is built on Matplotlib and simplifies working with a broader range of complex datasets, especially for statistical plotting.

### Common Seaborn Functions for Plot Customization

+ **hue** - Differentiates data categories using distinct colors.
+ **palette** - Defines the color scheme for categorical data.
+ **color** - Sets a single color for the entire plot line.
+ **style** - Modifies the style of plot elements (e.g., line style, marker style).
+ **markers** - Adds distinct markers to data points in the plot.
+ **legend** - Configures the legend's display or visibility.
+ **size** - Adjusts the line thickness based on a numerical variable.
+ **linewidth** - Specifies the width of the line.
+ **linecolor** - Sets the color of the line explicitly.
+ **saturation** - Adjusts the intensity of colors for filled elements
+ **alpha** - Controls the transparency level of the line or markers.
+ **orient** - Specifies the orientation of the plot, either "v" for vertical or "h" for horizontal.
+ **order** - Defines the specific order of categories to display on the categorical axis.
+ **hue_order** - Sets the order of categories for the hue variable in multi-color plots.
+ **xticklabels** - customize the labels displayed on the x-axis.
+ **yticklabels** - customize the labels displayed on the y-axis.

### Importing seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Line Plot

A Seaborn **line plot** is a graphical representation of data points connected by straight lines.

```python
k1 = [1,6,3,8,5,2,4]
k2 = [8,3,4,1,3,6,5]

sns.lineplot(x=k1, y=k2, color="g")

plt.show()
```

```python
k3 = [1,2,3,4,5,6,7]
k4 = [2,3,4,1,5,6,7]

x1 = pd.DataFrame({"k3":k3, "k4":k4})
sns.lineplot(x="k3", y="k4", data=x1, color="r")

plt.show()
```

```python
k5 = sns.load_dataset("tips").head(25)
k5
```

```python
sns.lineplot(x="tip", y="total_bill", data=k5, color="m")
plt.show()
```

```python
sns.lineplot(x="tip", y="total_bill", data=k5, hue="sex", style="sex", palette="hsv", markers=["o","<"], 
             dashes=False,legend=False)

plt.grid()
plt.show()
```

## Histogram Plot

A **histogram** is a graphical representation of the distribution of numerical data. It divides the data into bins (or intervals) and then counts the number of data points that fall within each bin. The height of each bar in the histogram corresponds to the count of data points in that bin.

+ **bins** - Intervals into which data is grouped in a histogram.
+ **kde** - **(Kernel Density Estimate)** A smooth curve that estimates the probability density function of a dataset.
+ **rug** - A plot of short vertical lines at each data point, used to visualize data density.

```python
k6 = sns.load_dataset("geyser")
k6
```

```python
sns.displot(k6["waiting"], color="g")
plt.show()
```

```python
sns.displot(k6["waiting"], bins=[40,50,60,70,80,90,100], kde=True, rug=True,color="r")
plt.show()
```

## Bar Plot

A **bar plot** is a categorical plot in Seaborn that represents data as rectangular bars, with the length of each bar proportional to the value of the corresponding category.

+ **capsize** - Specifies the size of the caps at the ends of error bars in a bar plot.
+ **err_kws** - Allows customization of error bars (e.g., {"linewidth": 2, "color": "r"}).

```python
k7 = sns.load_dataset("titanic")
k7
```

```python
sns.barplot(x=k7.embark_town, y=k7.survived, color="y")
plt.show()
```

```python
order_1 = ["Queenstown", "Southampton", "Cherbourg"]
sns.barplot(x="embark_town", y="survived", data=k7, hue="sex", order=order_1, hue_order=["female","male"],
            errorbar=('ci', 50), palette="turbo")
plt.show()
```

```python
sns.barplot(x=k7.age, y=k7.survived, orient="h", color="b", saturation=0.5, capsize=0.2, alpha=0.5, err_kws={'color': 'g', 'linewidth': 5})
plt.show()
```

## Scatter Plot

A **scatter plot** displays individual data points to visualize the relationship between two numerical variables.

```python
k8 = sns.load_dataset("iris").head(100)
k8
```

```python
sns.scatterplot(x="sepal_length", y="sepal_width", data=k8, color="g")
plt.show()
```

```python
m = {"setosa":"*","versicolor":"o"}

sns.scatterplot(x="sepal_length", y="sepal_width", data=k8, hue="species", style="species", size="species", sizes=(100,60),
                palette="ocean", alpha=0.5, markers=m)
plt.show()
```

## Heatmap

A **heatmap** is a powerful visualization tool that uses color to represent data values in a matrix.

+ **vmin** - Sets the minimum value for the color scale.
+ **vmax** - Sets the maximum value for the color scale.
+ **cmap** - Stands for colormap. It defines the color scheme used in the heatmap (ie..viridis, coolwarm, plasma, magma).
+ **annot** - Short for annotation. If True, the heatmap will display the data values inside the cells.
+ **cbar** - Refers to the color bar displayed alongside the heatmap. If True, a color bar is shown; if False, it is hidden.

```python
k9 = np.linspace(1,10,20).reshape(4,5)
k9
```

```python
sns.heatmap(k9)
plt.show()
```

```python
k10 = sns.load_dataset("anagrams")
k10
```

```python
k10 = sns.load_dataset("anagrams")
k11 = k10.drop(columns=["subidr","attnr"]).head(10)
k11
```

```python
sns.heatmap(k11, vmin=0, vmax=10, cmap="mako", annot=True, linewidth=0.5, linecolor="g", cbar=True, xticklabels=True, yticklabels=False)
plt.show()
```

## Count Plot

A **count plot** is a bar plot used to show the frequency of occurrences for each category in a dataset.

```python
k12 = sns.load_dataset("mpg").head(100)
k12
```

```python
sns.countplot(x="cylinders", data=k12, color="g")
plt.show()
```

```python
sns.countplot(x="cylinders", data=k12, hue="origin", palette="bone", saturation=1)
plt.show()
```

## Violine Plot

A **violin plot** in Seaborn is used to visualize the distribution of numerical data across different categories. It is a combination of a box plot and a kernel density estimate (KDE), which shows both summary statistics and the distribution shape of the data.

+ **split** - Determines whether to separate the violins for different categories when a hue variable is specified.
+ **inner** - Controls the display of inner markings within the violin plot.

```python
k13 = sns.load_dataset("penguins")
k13
```

```python
sns.violinplot(y=k13["body_mass_g"], color="c")
plt.show()
```

```python
sns.violinplot(x="species", y="bill_length_mm", data=k13, color="y")
plt.show()
```

```python
sns.violinplot(x="species", y="bill_length_mm", data=k13, hue="sex", linewidth=1, palette="Accent", order=["Gentoo", "Adelie", "Chinstrap"],
              saturation=1, split=True, inner="quart")
plt.show()
```

## Pair Plot

A **pair plot** is used to explore relationships between multiple variables in a dataset by plotting pairwise relationships between them.

+ **kinds** - The type of plot for off-diagonal relationships.

```python
k14 = sns.load_dataset("fmri")
k14
```

```python
sns.pairplot(k14)
plt.show()
```

```python
sns.pairplot(k14, hue="region", palette="CMRmap", kind="reg")
plt.show()
```

## Strip Plot

A **strip plot** is a type of scatter plot in Seaborn that displays individual data points along a categorical axis. It is useful for visualizing the distribution of data points for different categories.

+ **jitter** - Jitter refers to adding random noise to the position of points along the categorical axis to reduce overlap and make the distribution of points clearer.

```python
k15 = sns.load_dataset("dots")
k15
```

```python
sns.stripplot(y=k15["firing_rate"], color="y")
plt.show()
```

```python
sns.stripplot(x="choice", y="firing_rate", data=k15, color="g")
plt.show()
```

```python
sns.stripplot(x="choice", y="firing_rate", data=k15, hue="align", palette="gist_rainbow", linewidth=0.2, jitter=1, size=5, marker="<", alpha=1)
plt.show()
```

## Box Plot

A **box plot** shows the distribution of data using minimum, Q1, median, Q3, and maximum, with outliers and whiskers.

+ **showmeans** - Displays the mean as a marker on the box plot when set to True.
+ **meanprops** - Customizes the appearance of the mean marker, like its shape, color, and size

```python
k16 = sns.load_dataset("exercise").head(100)
k16
```

```python
sns.boxplot(y=k16["pulse"], color="y")
plt.show()
```

```python
sns.boxplot(x="kind", y="pulse", data=k16, color="c")
plt.show()
```

```python
sns.boxplot(x="kind", y="pulse", data=k16, color="g", order=["running","walking", "rest"], showmeans=True, 
            meanprops={"marker":"+", "markeredgecolor":"y"}, linewidth=1, orient="v")
plt.show()
```

## Cat plot

A **cat plot** in Seaborn is a versatile function to create categorical plots, combining several types of plots (strip, swarm, box, violin, bar, count, etc.) into one interface.

+ **kind** - kind parameter specifies the type of categorical plot to create.

```python
k17 = sns.load_dataset("taxis")
k18 = k17.drop(columns=["pickup","dropoff","pickup_zone","dropoff_zone","pickup_borough","dropoff_borough","color"]).head(100)
k18
```

```python
sns.catplot(x="passengers", y="total", data=k18, color="b")
plt.show()
```

```python
sns.catplot(x="passengers", y="total", data=k18, hue="payment", palette="afmhot", height=5)
plt.show()
```

```python
sns.catplot(x="passengers", y="total", data=k18, hue="payment", palette="BuGn", height=5, kind="bar")
plt.show()
```

```python
sns.catplot(x="passengers", y="total", data=k18, hue="payment", palette="gist_ncar", height=4, kind="point")
plt.show()
```

## Multi plot

A **multi-plot** refers to a collection of subplots (multiple plots) displayed together in a single figure, allowing for comparisons and analyses of different datasets or different views of the same data.

+ **FacetGrid** - A FacetGrid is a powerful tool in Seaborn for creating multi-plot grids. It allows you to visualize subsets of your data by mapping variables to rows, columns, or hue (color). This is particularly useful for comparing distributions or relationships across categories.

```python
k20 = sns.load_dataset("diamonds").head(50)
k20
```

```python
ok = sns.FacetGrid(k20, col="cut")
ok.map(plt.scatter, "x", "y")
plt.show()
```

## Styling plot

+ **sns.set(style=" ")** -Sets the overall aesthetic style of the plot. Available styles: ("darkgrid", "whitegrid", "dark", "white", and "ticks")
+ **plt.figure(figsize=(10,4))** - Creates a new figure with a specified size of 10 inches in width and 4 inches in height.
+ **sns.set_context()** - Adjusts the size and scale of elements like labels, titles, and ticks for a specific context.
+ **sns.despine()** - Removes the top and right spines (lines surrounding the plot), giving the plot a cleaner look.

```python
k19 = sns.load_dataset("flights").head(20)
k19
```

```python
sns.set(style="white")

plt.figure(figsize=(8,4))

sns.set_context("poster",font_scale=0.5)

sns.barplot(x="month", y="passengers", data=k19, hue="year", palette="Accent")
sns.despine()

plt.show()
```
