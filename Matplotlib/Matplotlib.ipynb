### Data Visualization

**What is Data Visualization ?**

Data visualization is the graphical representation of information and data. It uses visual elements like charts, graphs, and maps to make complex data more understandable and accessible. By turning raw data into visual stories, data visualization helps people to identify patterns, trends, and outliers that might be difficult to spot in a spreadsheet or other text-based format.

**Key benefits of Data Visualization :**

+ Improved understanding
+ Enhanced decision-making
+ Effective communication
+ Storytelling

**Tools for Data Visualization :**

+ **Software** : Tableau, Power BI, Excel, Python libraries (***Matplotlib***, Seaborn, Plotly, ggplot).
+ **Online platforms** : Google Data Studio, Infogram, Canva.

### Importing Matplotlib 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### ***Linear Plot***

+ **Linear plot :** A linear plot is a graphical representation of a linear equation or relationship between two variables. It is typically displayed as a straight line on a coordinate plane.

```python
x = [1,2,3,4]
y = [4,5,6,7]
plt.plot(x,y,color='y')
plt.show()
```

### ***Bar Plot***

+ **Bar plot :** A bar plot is type of plot used to represent categorical data. It displays data as rectangular bars, where the length or height of each bar corresponds to the value it represents.

    + ***width :*** Controls the width of the line.
    + ***align :*** Determines the alignment of the line markers relative to the data points.
    + ***edgecolor :*** Controls the color of the edges of the bars.
    + ***linewidth :*** Controls the width of the edges of the bars.
    + ***linestyle :*** controls the style of the edges of the bars.
    + ***alpha :*** Controls the transparency of the bars.

```python
x = ["Omi", "Krishna", "Aditi", "Mukul", "Anushka"]
y = [21, 24, 21, 24, 19]

plt.xlabel("Name", fontsize=15)
plt.ylabel("Age", fontsize=15)
plt.title("Friends Age", fontsize=15)

c = ["r", "y", "b", "g", "m"]
plt.bar(x,y,width=0.5,color=c, align="center", edgecolor="k", linewidth=1, linestyle="-", alpha=0.8, label="Krishna")

plt.legend()
plt.show()
```

```python
x = ["Omi", "Krishna", "Aditi", "Mukul", "Anushka"]
y = [21, 24, 21, 24, 19]
z = [49, 30, 32, 45, 20]

width=0.3
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("Name", fontsize=15)
plt.title("Friends", fontsize=15)
plt.bar(p,y,width,color='g',label='Age')
plt.bar(p1,z,width,color='y',label='Marks')

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

### ***Scatter Plot***

+ **Scatter plot :** A Scatter plot is a type of plot that displays the relationship between two sets of data as points on a grid.

    + ***size :*** Controls the size of each data point.
    + ***marker :*** Determines the shape of the data points (eg.. 'o', '*').
    + ***cmap :*** Assigns colors to the data points based on the colormap.

+ **Colorbar :** A colorbar is a visual scale that associates colors with specific numerical values or categories.

```python
x = [1,2,3,4,5,6]
y = [6,5,4,3,2,1]

size=[200,300,400,500,100,250]
colors=[64,24,84,12,96,50]

plt.title("Scatter Plot",fontsize=14)
plt.scatter(x,y,c=colors,s=size,marker='*',cmap="BrBG")

t=plt.colorbar()
t.set_label("ColorBar")

plt.show()
```

### ***Histogram Plot***

+ **Histogram plot :** It is a statistical visualization tool used to represent the distribution of a numerical dataset. It shows the frequency or density of data points within the certain intervals or bins.

    + ***bins :*** Controls the number of bins (interval) used to group the data in the histogram.
    + ***rwidth :*** Controls the relative width of the bars in the histogram.
    + ***axvline() :*** Add a line to the plot.
    + ***bottom :*** Specifies the baseline for the histogram bars.
    + ***histype :*** Controls the type of histogram to be plotted (eg.. 'bar', 'step', 'fill', 'stepfilled').
    + ***log :*** Use a logarithmic scale on the y-axis of a hitogram.

```python
num = [42, 34, 84, 46, 28, 45, 13, 87, 25, 43, 84, 23, 89, 29, 62, 34, 25, 74, 62, 82, 56, 24, 83, 73, 81, 73, 39]
l = [10,20,30,40,50,60,70,80,90]
plt.hist(num,bins=l,color='g',edgecolor='y',rwidth=0.99)
plt.axvline(50,color='r',label="line")
plt.legend()
plt.grid()
plt.show()
```

```python
num = [42, 34, 84, 46, 28, 45, 13, 87, 25, 43, 84, 23, 89, 29, 62, 34, 25, 74, 62, 82, 56, 24, 83, 73, 81, 73, 39]
l = [10,20,30,40,50,60,70,80,90]
plt.hist(num,bins=l,edgecolor='y',bottom=10, histtype="step",orientation="horizontal",log=True)
plt.show()
```

### ***Pie Plot***

+ **Pie plot :** A pie plot is a circular chart divided into sectors, each representing a category or group of data. The size of each sector is proportional to the quantity of the category it represents.

    + ***explode :*** Creates an "exploded" effect where one or more slices of the pie chart are pulled away from the center.
    + ***autopct :*** Specifies the format for displaying percentages on each slice (eg.. ***%0.1f%%*** for percentage sign with one decimal place.
    + ***shadow :*** Add a shadow effect to the pie chart for a more 3D appearance.
    + ***radius :*** Sets the radius of the pie chart.
    + ***labeldistance :*** Controls the distance of the labels from the center of the pie chart.
    + ***startangle :*** Specifies the starting angle for the first slice.
    + ***textprops :*** Set the properties of the text labels (eg.. fontsize, color, fontweight).
    + ***wedgeprops :*** Sets the properties of the pie chart wedges (eg.. edgecolor, linewidth, alpha).
    + ***rotatelabels :*** Rotate the labels to avoid overlapping.

```python
x = [8, 5, 3, 5, 1, 2]
y = ["sleep", "study", "travel", "phone/laptop", "cook", "other"]

ex = [0.0,0.2,0.0,0.0,0.0,0.0]
c = ['y', 'g', 'c', 'r', 'm', 'b']

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1,labeldistance=1.1,startangle=220,
        textprops={"fontsize":10},wedgeprops={"linewidth":1,"edgecolor":'k'},rotatelabels=False)

plt.title("My Daily Plan")
plt.legend(loc=4)

plt.show()
```

### ***Stem Plot***

+ **Stem plot :** It displays data as stems extending from a baseline along a vertical or horizontal axis. Each stem represents a data point, with a line (the stem) connecting to the baseline to the data point, and a marker at the tip of the stem.

    + ***linefmt  :*** Controls the format of the lines connecting the data points to the baseline.
    + ***markerfmt :*** Control the format of the markers placed at the end of the lines.
    + ***basefmt :*** Controls the format of the base line.
    + ***bottom :*** Specify the y-value of the base line.
    + ***orientation :*** Specifies the orientation of the stem plot.

```python
x = [1,2,3,4,5,6]
y = [2,3,2,5,4,3]

plt.stem(x,y,linefmt=":",markerfmt="yo",bottom=0,basefmt="g",orientation="vertical")

plt.show()
```

### ***Box Plot***

+ **Box plot :** A box plot, also known as box and whisker plot, is a statistical chart used to visualize the distrubation of a dataset. It provides a summary of the data's quartiles, median, and potential outliers.

    + ***notch :*** Create a notch around the median of each box plot.
    + ***vert :*** Specifies the orientation of the box plots.
    + ***widths :*** Set the width of each box plot.
    + ***labels :*** Provide labels for each box plot.
    + ***patch_artist :*** Controls whether the box plots are treated as artists.
    + ***showmeans :*** Displays the mean of each box plot.
    + ***sym :*** Specifies the symbol used to represent outliers.
    + ***boxprops :*** Set the properties of the boxes.
    + ***capprops :*** Set the properties of the caps.
    + ***whiskerprops :*** Set the properties of the whiskers.


##### **Key components of a box plot -**

**1.Median :** The middle value of the dataset when arranged in ascending order. It's represented by a line within the box.

**2.Quartiles :**

+ ***First Quartile (Q1) -*** The middle value of the lower half of the data.
+ ***Second Quartile (Q3) -*** The middle value of the higher value of the data.
+ ***Interquartile Range (IQR) -*** The difference between Q3 and Q1, representing the spread of the middle 50% of the data.

**3.Whiskers :** Lines extending from the box to the minimum and maximum values within a certain range.

**4.Outliers :** Data points that fall outside the whiskers, potentially indicating unusual or extreme values.

```python
x = [10,20,30,40,50]

plt.boxplot(x)

plt.show()
```

```python
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,notch=True,vert=False,widths=0.3,labels=["Python"],patch_artist=True,showmeans=True,sym="r+",boxprops=dict(color="y"),
            capprops=dict(color="b"),whiskerprops=dict(color="g"))

plt.show()
```

**Multiple Box Plot**

```python
x = [10,20,30,40,50,60,70]
y = [10,20,40,50,30,70,80]
z = [x,y]

plt.boxplot(z,showmeans=True,sym="r+",boxprops=dict(color="y"),capprops=dict(color="b"),whiskerprops=dict(color="g"))

plt.show()
```

### ***Stack Plot***

+ **Stack plot :** A stack plot is a type of plot that diplays the contribution of each category to the total value across different categories. It is a great way to visualize how different components add up to a whole. Stack plots are often based on area charts, where each area represents a different components.

          ie.. A stack plot  is like slicing the cake into different pieces. Each slice is a different color, and they are stacked on top of each other. The whole cake is the total, and each slice shows how much that category contributes to the total.

```python
x = [1,2,3,4,5]

area1 = [3,1,2,2,5]
area2 = [4,5,2,6,3]
area3 = [2,1,5,1,4]

l = ["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["y","r","g"],baseline="zero")
plt.legend(loc=2)
plt.show()
```

### ***Step Plot***

+ **Step plot :** Step plot are a type of line plot where the data points are connected by horizontal lines, creating a staircase-like effect.

    + ***marker :*** The marker parameter controls the type of marker used to represents the data points.
    + ***ms :*** ms parameter controls the size of the markers used to represent the data points.

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.step(x,y,color="g",marker="o",ms=5,mfc="r")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

### ***Fill-Between Plot***

+ **Fill between plot :** A fill between plot is a type of plot that fills the area between two horizontal curves.

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y,color="r")

plt.fill_between(x,y,color="y",alpha=0.5)

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

### xticks(), yticks(), xlim(), ylim() & axis()

+ **xticks() & yticks()**

    + ***xticks() :*** It customize the numbers and labels along x-axis.
    + ***yticks() :*** It customize the numbers and labels along y-axis.

```python
x = [1,2,3,4]
y = [21,24,21,24]

plt.plot(x,y,color="r")

plt.xlabel("Name")
plt.xticks(x,labels=["Omi","Krishna","Aditi","Mukul"])

plt.ylabel("Age")
plt.yticks(y)

plt.show()
```

+ **xlim() & ylim()**

    + ***xlim() :*** Sets the minimum and maximum values for x-axis.
    + ***ylim() :*** Sets the minimum and maximum values for y-axis.

```python
x = [1,2,3,4]
y = [21,24,21,24]

plt.plot(x,y,color="r")

plt.xlabel("Name")
plt.xlim(0,10)

plt.ylabel("Age")
plt.ylim(20,25)

plt.show()
```

+ **axis()**

    + ***axis() :*** It is used to set and get the x and y limits of the plot.

```python
x = [1,2,3,4]
y = [21,24,21,24]

plt.plot(x,y,color="r")

plt.xlabel("Name")
plt.ylabel("Age")

plt.axis([0,10,15,25])

plt.show()
```

### Working with text

+ **plt.text() :** It allows us to add text annotations to our plots. It offers flexibility in customizing the appearance and placement of text elements, making it essential for labeling, annotating and adding informative details to your visualizations.
  
    + ***style :*** Select the font style (eg.. italic, normal).
    + ***bbox :*** Creates a bounding box around the text with customizable properties (eg..color, alpha, padding).
    + ***rotation :*** Rotate the text with specified angle (in degrees).

+ **plt.annotate() :** It allows us to add annotations to your plots. It offers flexibility in customizing the appearance and placement of annotations, making it an essential tool for enhancing the clarity and informativeness of your visualizations.

    + ***xy :*** Coordinates specifying the position of the annotation's arrowhead. 
    + ***xytext :*** Coordinates specifying the position of the text label.
    + ***arrowprops :*** A dictionary specifying the properties of the arrow connecting the annotation text to the arrowhead.

```python
x = [1,2,3,4,5,6,7]
y = [5,4,4,6,3,2,1]

plt.plot(x,y,color="b")

plt.title("My Study Routine",fontsize=14)

plt.xlabel("Days")
plt.ylabel("Study hours")

plt.text(1,1,"This graph shows, I'm not consistent.",fontsize=9,style="italic",bbox={"facecolor":"yellow"},rotation=0)

plt.annotate("See Here",xy=(2.5,1.5),xytext=(3,3),arrowprops=dict(facecolor="black"))

plt.show()
```
